**TEST DATASET**

```python
testdf=pd.read_csv("/kaggle/input/titanic/test.csv")
testdf
```

```python
testdf.isnull().sum()
```

```python
testdf=testdf.drop("Age",axis=1)
```

```python
testdf.isnull().sum()
```

```python
testdf=testdf.drop(["Name","Cabin","Ticket"],axis=1)
```

```python
testdf.info()
```

```python
X_train=traindf["Survived"]
X_test=traindf.drop("Survived",axis=1)
```

```python
y = traindf["Survived"]

features = ["Pclass", "Sex", "SibSp", "Parch"]
x = pd.get_dummies(traindf[features])
X_test = pd.get_dummies(testdf[features])
```

```python
# importing random forest classifier from assemble module
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
# creating dataframe of IRIS dataset
rf_classifier = RandomForestClassifier(n_estimators=165, random_state=62,max_depth=5)
rf_classifier.fit(x, y)
```

```python

predictions = rf_classifier.predict(X_test)

output = pd.DataFrame({'PassengerId': testdf.PassengerId, 'Survived': predictions})
output.to_csv('submission.csv', index=False)
print("Your submission was successfully saved!")
```
